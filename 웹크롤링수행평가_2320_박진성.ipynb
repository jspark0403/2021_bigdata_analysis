### 필요한 라이브러리 불러옴

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import time
```

### 크롤링할 Url 가져오기, 요청해보기

```python
url = "https://finance.naver.com/sise/sise_market_sum.nhn"
res = requests.get(url)
res
```

```python
res.text
```

```python
soup = BeautifulSoup(res.text)
```

## 종목명 추출

### 종목명을 for문으로 받아오기

```python
name = soup.select('.tltle')
names=[]
for tag in name:
    names.append(tag.get_text().strip())
NAMES = np.reshape(names,(50,1))
data1 = pd.DataFrame(NAMES)
```

### 다른 값들도 for문으로 받아오기

```python
number = soup.select('.number')
numbers=[]
for tag in number:
    numbers.append(tag.get_text().strip())
NUMBER = np.reshape(numbers,(50,10))
data2 = pd.DataFrame(NUMBER)
```

### DataFrame합치기, 인덱스 +1

```python
STOCK = pd.concat([data1, data2], axis = 1)
STOCK.index += 1
STOCK
```

### 컬럼명 입력

```python
STOCK.columns = ['종목명', '현재가', '전일비', '등락률', '액면가', '시가총액', '상장주식수', '외국인비율', '거래량', 'PER', 'ROE' ]
```
