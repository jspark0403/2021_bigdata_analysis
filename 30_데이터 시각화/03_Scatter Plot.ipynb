# 3. Scatter Plot
* 좌표상의 점들을 표시함으로써 두 변수간의 관계를 나타내줌

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.random.rand(50) # [0, 1]
y = np.random.rand(50)
plt.scatter(x, y)
plt.show()
```

```python
plt.figure(figsize = (10, 10))
plt.scatter(x, y)
x = np.random.rand(50) # [0, 1]
y = np.random.rand(50)
plt.scatter(x, y)
plt.show()
```
